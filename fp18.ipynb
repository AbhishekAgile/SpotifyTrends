```python
import pandas as pd
import numpy as np
import pyfpgrowth
```

```python
df1 = pd.read_csv("top10x.csv")
```

```python
records = []
for i in range(0,200):
    records.append([str(df1.values[i,j])for j in range (1,9)])
```

```python
patterns = pyfpgrowth.find_frequent_patterns(records, 2)
```

```python
print(len(patterns))
```

```python
rules = pyfpgrowth.generate_association_rules(patterns,0.2)
```

```python
print(len(rules))
```
